```python
import  matplotlib.pyplot as plt
```

```python
# Ejemplo de una linea simple
x = [0,1,2,3,4,5]
y = [0,1,2,3,4,5]

plt.plot(x,y)
plt.scatter(x,y)
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.title("Mi primera grafica")
plt.show()
```

```python
# Grafico de barras
categorias = ["A", "B", "C", "D"]
valores = [10,20,25,30]
plt.bar(categorias, valores)
plt.title("Grafico de barras")
plt.xlabel("Categorias")
plt.ylabel("Valores")
```
